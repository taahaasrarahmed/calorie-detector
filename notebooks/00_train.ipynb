```python
# src/00_train_food.py

from ultralytics import YOLO

# 1) Point to your dataset YAML (fix path if different)
DATA_YAML = "../data/data.yaml"   # your file shows train: train/images etc.

# 2) Pick a small model to fine-tune
model = YOLO("yolov8n.pt")     # nano: fastest/smallest

# 3) Quick smoke test (uncomment to try 1 epoch first)
#model.train(data=DATA_YAML, epochs=1, imgsz=640, batch=8, device=0)

# 4) Real training run
model.train(
    data=DATA_YAML,
    epochs=50,
    imgsz=640,
    batch=16,      # drop to 8 or 4 if GPU memory is low
    device=0,      # set to "cpu" if no GPU
    project="outputs",
    name="food7_yolov8n",
    patience=10    # early stopping
)

print("\nDone. Best weights at: outputs/food7_yolov8n/weights/best.pt")
```
